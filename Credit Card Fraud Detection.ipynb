Importing The Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load the data set to pandas dataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# print first five rows of datasets
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# checking the number of missing value in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction & fraud transaction
credit_card_data['Class'].value_counts()
```

0 for normal transaction
1 for fraudlant transaction

```python
# seperating data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statsical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under sample dataset containing similar distribution of normal transaction  and fraudlant
the no. of fraudlant trans. is 163

```python
legit_sample = legit.sample(n=163)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting data into features and taregts

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

to split data into train and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

model Training
Logisctic Regression

```python
model = LogisticRegression()
```

```python
# Training the logistc Regresssion model with Trainig data
model.fit(X_train, Y_train)
```

Accuracy score


```python
# accuracy on trainig data
X_training_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_training_prediction, Y_train)
```

```python
print('Accuracy on training data :', training_data_accuracy)
```

```python
# accuracy on Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test data:', test_data_accuracy)
```
